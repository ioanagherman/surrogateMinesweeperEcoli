```python
import pandas as pd
```

# Check smallest dividing segments from stage 3

```python
dividing_blue = pd.read_csv('/home/ig13470/Documents/MinesweeperEcoli/Minesweeper_1.0/OUTPUT_script_4X/dividing_blue_0.txt', sep='\t', header=None)

with open('/home/ig13470/Documents/MinesweeperEcoli/Minesweeper_0.9/OUTPUT_script_3/blue_0_combosandgenes.txt') as file:
    lines = file.readlines()

blue_combos = [line.strip() for i, line in enumerate(lines) if i % 2 == 0]
blue_genes = [line.strip() for i, line in enumerate(lines) if i % 2 != 0]

all_blue = pd.DataFrame({'combos': blue_combos, 'genes': blue_genes})
all_blue['# genes'] = all_blue['genes'].str.split(' ').str.len()
```

```python
pd.merge(dividing_blue, all_blue, left_on=1, right_on='combos').sort_values('# genes').tail(15)
```
